```python
import nltk
nltk.download('vader_lexicon')
```

```python
import sqlite3

# Create a new SQLite database
conn = sqlite3.connect('Converstaion.db')
c = conn.cursor()

# Create a table to store transcripts and analysis results
c.execute('''CREATE TABLE IF NOT EXISTS transcripts
             (id INTEGER PRIMARY KEY AUTOINCREMENT, transcript TEXT, sentiment_score REAL)''')
conn.commit()
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

# Initialize the sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Example transcript
transcript = input()

# Perform sentiment analysis
scores = sia.polarity_scores(transcript)
sentiment_score = scores['compound']

# Print the sentiment score
print(f"Sentiment Score: {sentiment_score}")

# Store the transcript and sentiment score in the database
c.execute("INSERT INTO transcripts (transcript, sentiment_score) VALUES (?, ?)", (transcript, sentiment_score))
conn.commit()

# Close the database connection
conn.close()
```
